# Q.3 List all the Movies in a Leap Year where Genre is Comedy

```python
#Comedy movies leap year

t4  = '''WITH Comedy_Movies AS
    (
        SELECT
             
            MG.MID
        FROM
                    GENRE G
            JOIN    M_GENRE MG
            ON      G.GID = MG.GID
        WHERE
            TRIM(G.Name) = 'Comedy'
    ),
    Comedy_Movies_In_Leap_Yr AS
    (
        SELECT
            M.MID,
            M.title,
            CAST(SUBSTR(M.year,-4) AS INTEGER) year
        FROM
                    Comedy_Movies CM
            JOIN    Movie M
            ON      CM.MID = M.MID
        WHERE
             (( year%4 = 0  AND  year % 100 <> 0) OR  year % 400 = 0)
       
    )
    
    SELECT * FROM Comedy_Movies_In_Leap_Yr
                 
                 
'''
mdp_results = pd.read_sql_query(t4,conn)
print(mdp_results.shape)
print(mdp_results.head(10))
```

```python
%%time
def grader_3(q3):
    q3_results  = pd.read_sql_query(q3,conn)
    print(q3_results.head(10))
    print(q3_results.shape)
    assert (q3_results.shape == (300,1))

query3 = '''WITH
    ACTORS_BEFORE_1970 AS
    (
        SELECT
            DISTINCT
            TRIM(MC.PID) PID
        FROM
                    Movie M
            JOIN    M_Cast MC
            ON      M.MID = MC.MID
        WHERE
            CAST(SUBSTR(M.year,-4) AS INTEGER) < 1970
    ),
    ACTORS_AFTER_1990 AS
    (
        SELECT
            DISTINCT
            TRIM(MC.PID) PID
        FROM
                    Movie M
            JOIN    M_Cast MC
            ON      M.MID = MC.MID
        WHERE
            CAST(SUBSTR(M.year,-4) AS INTEGER) > 1990
    )
    SELECT
        DISTINCT
        TRIM(P.Name) Actor_Name
    FROM
                ACTORS_BEFORE_1970 A_1970
        JOIN    ACTORS_AFTER_1990 A_1990
        ON      A_1970.PID = A_1990.PID
        JOIN    Person P
        ON      A_1970.PID = TRIM(P.PID)
'''

grader_3(query3)
```

# Q.11 List the Year of the movies that only have Female actors in it

```python
%%time
def grader_5a(q5a):
    q5a_results  = pd.read_sql_query(q5a,conn)
    print(q5a_results.head(10))
    print(q5a_results.shape)

    assert (q5a_results.shape == (78,2))
    
query5a = """

WITH
    movies_with_no_females as
    (
        SELECT
            distinct
            trim(mc.mid) mid
        FROM
                    m_cast mc
            JOIN    person p
            ON      trim(mc.pid) = trim(p.pid)
        WHERE
            trim(p.gender) in ('male','none') -- assuming  none as not female.
    )
 
 SELECT
        cast(substr(m.year,-4) as unsigned) year,
        count(distinct trim(mid) ) movies_with_only_females
   FROM
        movie m
    WHERE
        trim(mid) not in (select mid from movies_with_no_females)
    group by 
        cast(substr(m.year,-4) as unsigned)
    order by
        year

"""
grader_5a(query5a)
```

## Q12 --- Find all the actors that made more movies with Yash Chopra than any other director.

```python
%%time
def grader_8a(q8a):
    q8a_results  = pd.read_sql_query(q8a,conn)
    print(q8a_results.head(10))
    print(q8a_results.shape)
    assert (q8a_results.shape == (430, 3))

query8a ='''   
       SELECT
            trim(mc.pid) actor_pid,
            trim(md.pid) director_pid,
            count(distinct trim(md.mid)) as MOVIE_COUNT
        FROM
            m_cast mc,
            m_director md
        WHERE
            trim(mc.mid)= trim(md.mid)
            AND md.pid like '%nm0007181%' --ID for Yash Chopra = nm0007181
            
        GROUP BY
            actor_pid,
            director_pid
        
        ORDER BY MOVIE_COUNT DESC    


'''
grader_8a(query8a)

# using the above query, you can write the answer to the given question
```
